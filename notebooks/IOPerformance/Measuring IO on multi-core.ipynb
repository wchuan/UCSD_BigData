```python
def measureRandomAccess(size,filename='',k=100000,duty_cycle=0.01,rest_cycle=0.01,Seed=0):
    """ Measure the distribution of random accesses in computer memory.
    size=size of memory block (If reading from memory, otherwise it is the size of the file)
    filename= a file that is used as an external buffer. If filename=='' then everything is done in memory.
    k = number of times that the experiment is repeated.
    output:
    mean = the mean of T
    std = the std of T
    T = a list the contains the times of all k experiments
    cycle = defines the length of the duty cycle for this thread
    """
    import time,os
    import numpy as np
    from numpy.random import rand,seed
    from numpy import sqrt

    # Prepare buffer.
    if filename == '':
        inmem=True
        A=bytearray(size)
        #print 'reading memory, size=',size
    else:
        inmem=False
        size=os.stat(filename).st_size
        file=open(filename,'r+')
        #print 'reading file',filename,', size=',size
        
    # Read and write k times from/to buffer.
    sum=0; sum2=0
    readWriteSize=1000
    T=np.zeros(k)
    t_cycle=time.time()
    seed(100*Seed)
    for i in range(k):
        t=time.time()
        T[i]=t
        loc=int(rand()*size)
        if inmem:
            x=A[loc:loc+4]
            A[loc]=(i % 256)
        else:
            file.seek(loc)
            poke=file.read(readWriteSize)
            file.write('0123456789'*(readWriteSize/10))
        d=time.time()-t
        sum += d
        sum2 += d*d
        if time.time()-t_cycle>duty_cycle:
            time.sleep(rest_cycle)
            t_cycle=time.time()
    mean=sum/k; var=(sum2/k)-mean**2; std=sqrt(var)
    return (mean,std,T)
```
